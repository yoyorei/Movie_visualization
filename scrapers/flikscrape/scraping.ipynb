```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os
import csv
import requests
```

```python
# Files to Load
top_movies = os.path.join("Flickmetrix_topmovies.csv")
```

```python
movies_df = pd.read_csv(top_movies)
```

```python
movies_df.head()
```

```python
# Nonbreaking space at end of titles; removed by regex
movies_df = movies_df.replace(' ','', regex=True)
```

```python
movies_df.head()
```

```python
movies_df.columns
```

```python
movies_df.count()
```

```python
# Extract "filmInfoContainer", "filmInfoContainer1", "filmInfoContainer2", "megascore", "Field1","Field1_link", "Field2", 
# extract "Field2_link", "Field3", "Field3_link","Field4","Field4_link", "Field5", "Field5_link" 

reduced_movies_df = movies_df.loc[:, ["filmInfoContainer", "filmInfoContainer1", "filmInfoContainer2",
                                    "megascore", "Field1","Field1_link", "Field2", "Field2_link", "Field3", "Field3_link","Field4","Field4_link", "Field5", "Field5_link"]]
reduced_movies_df
```

```python
# Rename columns
reduced_movies_df = reduced_movies_df.rename(columns={"filmInfoContainer": "Movie_title", "filmInfoContainer1": "Year", "filmInfoContainer2": "Genre", "megascore": "Average_Rating",
                                                      "Field1": "Tomatometer", "Field1_link": "RT_link","Field2": "Audience_Rating", "Field2_link": "RT_link","Field3": "IMDb", "Field3_link": "IMDb_link",
                                                      "Field4": "Letterboxd","Field4_link": "LB_link", "Field5": "Metascore", "Field5_link": "Ms_link"})
                                                                                                        
```

```python
reduced_movies_df.head()
```

```python
reduced_movies_df.to_csv("movies_df.csv")
```
